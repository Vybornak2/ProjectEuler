### Without optimized libraries

```python
import math


def find_prime(n):
    primes = [2]
    number = 3

    while len(primes) < n + 1:
        is_prime = True
        sqrt_number = int(math.sqrt(number)) + 1
        for prime in primes:
            if prime > sqrt_number:
                break
            if number % prime == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(number)
        number += 1

    return primes[-1]
```

```python
def main():
    return find_prime(10_000)
```

```python
%%timeit
main()
```
