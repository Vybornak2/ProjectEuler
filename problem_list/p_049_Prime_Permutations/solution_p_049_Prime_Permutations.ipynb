## Brute-force Solution

```python
import sympy as sp
from collections import defaultdict
from itertools import combinations


def main_bf():
    candidates = list(sp.primerange(1e3, 1e4))

    groups = defaultdict(list)
    for c in candidates:
        key = "".join(sorted(str(c)))
        group = groups[key]
        group.append(c)

    for key, group in groups.items():
        if len(group) < 3:
            continue
        if 1487 in group:
            continue

        group_set = set(group)

        # instead of checking all triplets, check all pairs (a,c) and see if (a+c)/2 is in the set
        for a, c in combinations(group, 2):
            mid = (a + c) / 2
            if mid.is_integer() and int(mid) in group_set:
                return f"{a}{int(mid)}{c}"

    return ""
```

```python
%%timeit
main_bf()
```
