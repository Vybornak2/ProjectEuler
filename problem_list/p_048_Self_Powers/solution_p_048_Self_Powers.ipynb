## Brute-force Solution

```python
def main_bf():
    total = 0
    for i in range(1, 1001):
        total += i**i

    return str(total)[-10:]
```

```python
%%timeit
main_bf()
```

## Optimized Solution

```python
def pow_mod(base, exp, mod):
    result = 1
    for _ in range(exp):
        result = (result * base) % mod
    return result
```

```python
def main_opt():
    MOD = 10**10
    total = 0
    for i in range(1, 1001):
        total += pow_mod(i, i, MOD)

    return str(total % MOD)
```
