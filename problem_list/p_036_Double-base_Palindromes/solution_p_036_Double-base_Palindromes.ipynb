## Brute-force Solution

```python
def is_palindrome(n):
    return str(n) == str(n)[::-1]


def to_binary(n):
    return bin(n)[2:]


def main():
    total = 0
    for i in range(1, 1_000_000, 2):
        if is_palindrome(i) and is_palindrome(to_binary(i)):
            total += i
    return total
```

```python
%%timeit
main()
```
