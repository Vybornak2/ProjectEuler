```python
import numpy as np

fibonacci_seq = [0, 1, 2]
value = fibonacci_seq[-1]

while value < 4 * 10**6:
    value = fibonacci_seq[-1] + fibonacci_seq[-2]
    fibonacci_seq.append(value)

fibonacci = np.array(fibonacci_seq)
even_fibonacci = fibonacci[fibonacci % 2 == 0]

print(f"The sum of the even-valued terms is {even_fibonacci.sum()}")
```
