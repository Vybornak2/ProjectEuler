## Brute-force Solution

```python
def get_triangle_numbers(upper_limit: int) -> set[int]:
    triangle_numbers = set()
    for n in range(1, upper_limit):
        t_n = n * (n + 1) // 2
        triangle_numbers.add(t_n)

    return triangle_numbers
```

```python
def letter_value(letter):
    return ord(letter.upper()) - ord("A") + 1


def get_word_value(word):
    return sum(letter_value(letter) for letter in word)
```

```python
for letter in "SKY":
    print(f"{letter}: {letter_value(letter)}")
```

```python
def read_words() -> list[str]:
    with open("words_042.txt", "r") as f:
        words = f.read().split(",")
        words = [word.strip('"') for word in words]
    return words
```

```python
def main_bf():
    triangle_numbers = get_triangle_numbers(50)
    t_count = 0
    for w in read_words():
        value = get_word_value(w)
        if value in triangle_numbers:
            t_count += 1

    return t_count
```

```python
%%timeit
main_bf()
```

```python
main_bf()
```

```python
def is_triangle_number(value: int) -> bool:
    # using the inverse formula to check if a number is triangular
    n = (-1 + (1 + 8 * value) ** 0.5) / 2
    return n.is_integer() and n > 0
```

```python
def main_opt():
    t_count = 0
    for w in read_words():
        value = get_word_value(w)
        if is_triangle_number(value):
            t_count += 1

    return t_count
```

```python
%%timeit
main_opt()
```
